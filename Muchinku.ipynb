## Manipulating DNA strings

```python
def longestcommonprefix(s1,s2):
    i = 0
    while i < len(s1) and i < len(s2) and s1[i] ==s2[i]:
        i=i+1
    return s1[:i]
longestcommonprefix('AATACTCACTC','AATAGAGCGACGGCACA')
```

```python
def match(s1,s2):
    if len(s1) != len(s2):
        return False
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            return False
    return True
match('ATCG','ATCG')
```

## Reverse compliment using a dictionary and a for loop

```python
compliment={'A':'T', 'T':'A', 'C':'G', 'G':'C'}
```

```python
type(compliment)
```

```python
def reverse(s):
    compliment={'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    t=''
    for base in s:
        t=compliment[base] + t
    return t
reverse('ATACTTATCATTACT')
```

## Downloading and parasing a fastq file

```python
!wget --no-check https://d28rh4a8wq0iu5.cloudfront.net/ads1/data/SRR835775_1.first1000.fastq
```

```python
def readfastq(filename):
    sequences=[]
    qualities=[]
    with open(filename) as fh:
        while True:
            fh.readline()
            seq=fh.readline().rstrip()
            fh.readline()
            qual=fh.readline().rstrip()
            if len(seq) ==0:
                break
            sequences.append(seq)
            qualities.append(qual)
    return sequences,qualities
seq,qual=readfastq('SRR835775_1.first1000.fastq')
```

```python
def phred33ToQ(qual):
    return ord(qual) -33
```

```python
phred33ToQ('@')
```

```python
def creatHist(qualities):
    hist=[0]*50
    for qual in qualities:
        for phred in qual:
            q=phred33ToQ(phred)
            hist[q] +=1
    return hist
h= creatHist(qual)
print(h)
```
